```python
import pandas as pd
last_year = pd.read_csv("result_dataset/world_data2020.csv")
curr_year = pd.read_csv("result_dataset/world_data.csv")
```

```python
last_year.head(3)
```

```python
curr_year.head(3)
```

Merging columns by values in the Country column.

```python
df = pd.merge(left=last_year, right=curr_year, on='Country')
df.head(5)
```

Selecting the 3 columns required for visualization where Total-2020 are results of the selected month in 2020, and Total results of the same month in 2021.

```python
df = df[['Country', 'Total-2020', 'Total']]
df.head(10)
```

Sorting by the current year.

```python
df = df.sort_values(['Total','Total-2020'], ascending=[False,False])
```

Selecting the first 50 results.

```python
df_50 = df.head(50)
df_50
```

```python
import matplotlib.pyplot as plt
import matplotlib.lines as mlines


# Func to draw line segment
def newline(p1, p2, color='black'):
    ax = plt.gca()
    l = mlines.Line2D([p1[0],p2[0]], [p1[1],p2[1]], color='skyblue')
    ax.add_line(l)
    return l

# Figure and Axes
fig, ax = plt.subplots(1,1,figsize=(10,14), facecolor='#f7f7f7', dpi= 80)

# Points
ax.scatter(y=df_50['Country'], x=df_50['Total-2020'], s=50, color='#a3c4dc', alpha=0.7)
ax.scatter(y=df_50['Country'], x=df_50['Total'], s=50, color='#0e668b', alpha=0.7)

plt.legend(['April 2020','April 2021'])


# Line Segments
for i, p1, p2 in zip(df_50['Country'], df_50['Total-2020'], df_50['Total']):
    newline([p1, i], [p2, i])
    ax.annotate(int(p2), xy=(p2+3, i))
    
# Decoration
ax.set_facecolor('#f7f7f7')
ax.set_title("Top 50 Countries by Search Requests in Fintech, April 2021", fontdict={'size':12, 'family':'Lato'})
ax.set(ylabel='Country', xlabel='Total Score')

plt.show()
```

```python
df['Change'] = df['Total'] - df['Total-2020']
```

```python
change_50 = df.sort_values('Change', ascending=False).head(50)
```

```python
import plotly.graph_objects as go

x = change_50['Change']
y = change_50['Country']
#y = world_data[world_data.columns[1:-1]]

fig = go.Figure(data=[go.Bar(
            x=x, y=y,
            text=x,
            textposition='outside',
            #color_continuous_scale='Blues',
            marker_color="#3C98CA", #color relevant to FNN
            orientation='h',
        )],)
fig.update_layout(margin=dict(l=190, r=20, t=60, b=15), 
                  plot_bgcolor="white",
                  paper_bgcolor="#E4E4E4",
                  yaxis={'categoryorder':'total ascending'},
                  barmode='group', height=1200,
                  title_text="Top {} Countries with the Largest Growth of Interest in Fintech Between April {} and April {}".format(50, 2020, 2021),
                  xaxis_title="Change in Scores",
                  yaxis_title="Country",
                  font=dict(family="Montserrat", size=13, color="black") #title font from FNN site
)


fig.show(renderer="svg", height=1200, width=980)
```
